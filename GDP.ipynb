```python
import os
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Handle date time conversions between pandas and matplotlib
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()

# Use white grid plot background from seaborn
sns.set(font_scale=1.5, style="whitegrid")
```

```python
aus = pd.read_csv('austria.csv')
euro = pd.read_csv('europ.csv')
```

```python
aus
```

```python
aus.info()
```

```python
aus.describe()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax = sns.lineplot(euro.TIME,euro.Value,data=euro, color="red", label="Europian")
plt.ylabel('y-axis')
plt.xticks(rotation=90);
```

```python
fig, ax = plt.subplots(figsize=(20,10))
ax = sns.lineplot(aus.TIME,aus.Value,data=aus, color="blue", label="Austria")
plt.ylabel('y-axis')
plt.xticks(rotation=90);
```

```python
fig, ax = plt.subplots(figsize=(20,5))
ax = sns.lineplot(aus.TIME,aus.Value,data=aus, color="blue", label="Austria")
ax = sns.lineplot(euro.TIME,euro.Value,data=euro, color="red", label="Europian")
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax=sns.scatterplot(x=aus.TIME,y=aus.Value, data=aus, size="Value",sizes=(1,400));
```

```python
sw = pd.pivot_table(aus, 
                    index=['TIME'], 
                    columns=['UNIT'], 
                    values='Value')
```

```python
sw
```

```python
plt.figure(figsize=(8,8))
sns.heatmap(sw, annot=True, cmap='coolwarm').set_title('Seaborn');
```

```python
plt.figure(figsize=(15,10))
sns.residplot(x=sw['annualized percentage change on previous period'], y=sw['percentage change compared to same period in previous year'], data=sw,  color='g');
```
